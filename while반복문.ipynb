```python
# while 반복문은 조건식으로만 동작하여 반복할 코드 안에 조건식에 영향을 주는 변화식이 들어감

i = 0                       # 초기식
while i < 100:              # while 조건식
    print('Hello, world!')      # 반복할 코드
    i += 1                      # 변화식
```

```python
i = 0
while i < 100: # 변수 i가 100보다 작을 때까지 반복함
    print('Hello, world!')
    i += 1
```

```python
# 초기값을 1부터 시작하기

i = 1
while i <= 100:
    print('Hello, world!', i)
    i += 1
```

```python
i = 1
while i <= 100:
    print('Hello, world!', i)
    i += 2
```

```python
# 짝수 출력하기

i = 2
while i <= 100:
    print('Hello, world!', i)
    i += 2
```

```python
# 변수값을 감소시키기

i = 100
while i > 0: # i가 0보다 클 때까지 반복함 
    print('Hello, world!', i)
    i -= 1
```

```python
# 입력한 횟수대로 반복하기

count = int(input('반복할 횟수를 입력하세요: ')) # 정수형으로 변경

i = 0
while i < count: # i가 count보다 작을 때 반복
    print('Hello, world!', i)
    i += 1 # 1씩 증가시킨다
```

```python
count = int(input('반복할 횟수를 적어주세요: '))

while count >0:
    print('Hello, world!')
    count -= 1 # 1씩 감소시킨다
```

# 반복 횟수가 정해지지 않은 경우


```python
# 1과 6 사이의 난수를 생성한 뒤 3이 나오면 반복을 끝냄

import random # random 모듈을 가져옴

i = 0
while i != 3: # 3이 아닐때 계속 반복
    i = random. randint(1, 4) # randint를 사용해 1과 4 사이의 난수를 생성
    print(i)
```

# while 반복문으로 무한 루프 만들기

```python
while True: # while에 True를 지정하면 무한 루프
    print('Hello, world!')
```

```python
while 1: # 0이 아닌 숫자는 True로 취급하여 무한 루프로 동작
    print('Hello, world!')
```

```python
while 'Hello': # 내용이 있는 문자열은 True로 취급하여 무한 루프로 동작
    print('Hello, world!')
```
