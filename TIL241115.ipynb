API(Application Programming Interface): 프로그램이나 서비스 간의 소통을 위한 인터페이스

1. 호스트: 서버의 주소로, 일반적으로 웹사이트의 도메인과 동일하거나 API 전용 도메인을 가질 수 있다.
2. 포트 번호: 서버에서 API가 실행 중인 포트를 지정, 대부분의 경우 기본값(80 for HTTP, 443 for HTTPS)을 사용한다.
3. 경로: API에서 특정 리소스나 엔드포인트를 지정, 일반적으로 리소스에 대한 구조적 경로를 나타낸다.
4. 쿼리: 추가 데이터를 전달하기 위해 URL에 추가, ?로 시작하고 key=value 형태로 구성되며, 여러 개의 파라미터는 &로 구분한다.

오픈 API의 특징과 활용
1. 오픈 API는 API KEY를 활용해 사용자를 확인한다. 발급받은 API 키는 보안을 유지해야 한다.(타인의 악용을 방지)

2. 요청 제한을 확인한다. 일일/분당 호출 횟수 제한 확인(과도한 요청을 차단)

3. API 문서를 숙지한다. 요청 데이터와 응답 데이터 및 형식을 파악한다. 최신 사항과 변경 사항을 체크한다(버전의 호환성 문제)

4. 응답 받은 데이터를 저장한다. 적절한 형식으로 변환/저장하고, 수집된 데이터를 백업 관리한다.
