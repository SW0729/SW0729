변수
1. 변수 선언 및 할당

'=' 기호로 변수에 값을 할당할 수 있으며, 숫자로 시작할 수 없고 대소문자를 구분한다. 또한 파이썬의 예약어는 변수로 사용할 수 없다.

 

여러 변수에 동시에 값을 할당할 수 있다. (a, b, c = 1, 2, 3)

여러 변수에 같은 값을 할당하는 것 또한 가능하다. (x=y=z=100)

 

변수는 의미를 쉽게 알아볼 수 있도록 규칙에 알맞게 이름을 짓는 것이 좋다.

 

연산자
1. 산술 연산자

기본적인 수학 연산에 사용

+	덧셈
-	뺄셈
*	곱셈
/	나눗셈
%	나머지
**	거듭제곱
//	몫
 

2. 비교 연산자

두 값을 비교해 참(True)과 거짓(False)을 반환

==	값이 같음	5 == 5	True
!=	값이 다름	5 != 5	True
>	큼	5 > 3	True
<	작음	3 < 5	True
>=	크거나 같음	5 >= 5	True
=<	작거나 같음	4 =< 5	True
 

3. 본문 연산자

논리값(True, False)을 결합하여 새로운 논리값을 반환

And	두 조건 모두 참	True and False	False
Or	두 조건 중 하나가 참	True or False	True
Not	논리값을 반대로 뒤집음	not True	False
 

4. 대입 연산자

대입 연산자는 변수에 값을 할당할 때 사용

=	값 할당	x = 5	x = 5
+=	더한 후 할당	x += 3	x = x + 3
-=	뺀 후 할당	x -= 3	x = x - 3
*=	곱한 후 할당	x *= 2	x = x * 2
/=	나눈 후 할당	x /= 2	x = x / 2
%=	나머지를 구한 후 할당	x %= 2	x = x % 2
**=	거듭제곱 후 할당	x **= 2	x = x **= 2
//=	몫을 구한 후 할당	x //= 2	x = x //= 2
 

5. 비트 연산자

이진수(bit) 수준에서 연산을 수행

&	비트 AND	5 & 3	1
|	비트 OR	5 | 3	7
^	비트 XOR	5 ^ 3	6
~	비트 NOT(보수)	~5	-6
<<	왼쪽 시프트	5 << 1	10
>>	오른쪽 시프트	5 >> 1	2
 

6. 멤버십 연산자

특정 값이 시퀀스(문자열, 리스트, 튜플 등)에 속해 있는지 확인

in	시퀀스에 값이 포함되어 있는지 확인	"a" in "apple"	True
not in	시퀀스에 값이 포함되어 있지 않은지 확인	"b" not in "apple"	True
 

7. 식별 연산자

두 변수가 동일한 객체를 가리키는지 확인

is	두 변수가 동일 객체인지 확인	a is b	True or False
is not	두 변수가 동일하지 않은 객체인지 확인	a is not b	True or False
 

데이터 타입
숫자형은 정수형과 실수형으로 나뉜다.

 

1. 정수형: 소수점없이 정수를 나타내는 데이터 타입

2. 실수형: 소수점을 포함한 숫자를 나타내는 데이터 타입

3. 복소수형: 실수 부분과 허수 부분으로 구성된 숫자

 

문자형(String)은 문자, 단어, 문장을 저장하는 데이터 타입이다.

작은따옴표(') 또는 큰따옴표(")를 사용하여 표현할 수 있다.

 

문자형

name = "Alice"

greeting = "Hello, World!"

 

여러 줄 문자열(삼중 큰따옴표)

"""안녕하세요!

여러 줄의 문자열을 이렇게 작성 가능합니다."""

 

연결: +연산자로 문자열을 연결

반복: *연산자로 문자열을 반복

 

불리언

참 또는 거짓을 나타내는 데이터 타입으로 불리언 값은 주로 조건문에서 사용되며, True와 False 두 가지 값만 가질 수 있다.

 

불리언 연산

불리언 타입은 논리 연산자(and, or, not)와 함께 사용

 

a = True
b = False

print(a and b)  # False
print(a or b)   # True
print(not a)    # False

 

 

비교 연산의 결과

비교 연산자(==, != ,> ,< 등)를 사용한 결과는 불리언 값으로 반환

 

x = 10
y = 20

print(x == y)  # False
print(x < y)   # True

 

컬렉션 자료형

리스트(List): 순서가 있고 가변적인 자료형, 중복 허용.
튜플(Tuple): 순서가 있지만 불변적인 자료형, 중복 허용.
딕셔너리(Dictionary): 키-값 쌍으로 구성된 가변 자료형, 키는 유일하고 값은 중복 가능.
셋(Set): 중복되지 않는 요소들로 구성된 가변 자료형, 순서 없음.
 

제어문과 반복문

제어문: 프로그램이 결정을 내리도록 하는 도구로, 조건에 따라 코드를 실행하거나 실행하지 않도록 제어합니다.
반복문: 프로그램이 동일한 작업을 반복하도록 하는 도구로, 반복적인 작업을 자동화합니다.
결합 사용: 제어문과 반복문을 결합하여 복잡한 논리 흐름을 구성합니다.
조건문
If 문: 기본적인 조건문

if 뒤에 조건식이 오고, 조건식이 참이면 해당 코드 블록이 실행

 

age = 18

if age >= 18:
    print("성인입니다.")

 

Else 문: 조건이 거짓일 때 실행

else 문은 if 문의 조건이 거짓(False)일 때 실행되는 코드 블록을 정의

 

if 조건식:
    실행할 코드 (조건이 참일 때)
else:
    실행할 코드 (조건이 거짓일 때)

 

Elif 문: 여러 조건 검사하기

elif 문은 여러 개의 조건을 순차적으로 검사할 때 사용

if 문 다음에 여러 개의 elif 문을 사용하여 각 조건을 순서대로 검사하고, 처음으로 참인 조건이 발견되면 해당 코드 블록이 실행

 

if 조건식1:
    실행할 코드1
elif 조건식2:
    실행할 코드2
elif 조건식3:
    실행할 코드3
else:
    실행할 코드4 (모든 조건이 거짓일 때)

 

중첩 조건문 (Nested Conditions)

조건문 안에 또 다른 조건문을 사용할 수 있다. 이를 중첩 조건문(Nested Condition)이라고 한다.

 

if 조건식1:
    if 조건식2:
        실행할 코드1
    else:
        실행할 코드2
else:
    실행할 코드3

 

반복문
for 문: 컬렉션의 각 요소를 순회하면서 반복
while 문: 조건이 참인 동안만 반복
break 문: 반복문을 즉시 종료
continue 문: 현재 반복을 건너뛰고, 다음 반복을 수행
중첩 반복문: 반복문 안에 또 다른 반복문을 사용하여 복잡한 반복 작업을 수행
들여쓰기
들여쓰기란? Python에서 코드 블록을 구분하는 데 사용되는 중요한 요소로, 프로그램의 구조를 정의
들여쓰기 규칙: 일반적으로 공백 4칸을 사용하는 것이 표준이며, 탭과 공백을 혼용하지 않도록 주의
오류 방지: 잘못된 들여쓰기는 IndentationError를 발생시키므로, 일관된 들여쓰기를 유지하는 것이 중요
사용처: 조건문, 반복문, 함수 정의 등 모든 구조적 코드에서 들여쓰기가 필요
함수
함수란? 특정 작업을 수행하는 코드 블록으로, 이름을 통해 호출할 수 있으며 입력값을 받아 처리하고 결과값을 반환한다.
사용처: 반복되는 코드의 재사용, 코드 모듈화, 복잡한 작업의 분할 등 다양한 상황에서 사용
장점: 코드 재사용성, 가독성, 유지보수 용이성, 테스트 및 디버깅 용이성을 제공
주의사항: 함수는 단일 작업을 수행하도록 설계하며, 매개변수 수를 적절히 유지하고, 함수 이름은 기능을 명확히 표현해야 한다. 또한 부작용을 최소화해야 한다.
